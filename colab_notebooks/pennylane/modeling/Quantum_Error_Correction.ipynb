```python
def return_var_circ(num_wires):

  def circ_two():
    qml.Hadamard(wires = 0)
    for i in range(num_wires - 1):
      qml.CNOT(wires = [0, i + 1])
    return qml.state()

  device_two = qml.device("default.qubit", wires = num_wires)

  node_two = qml.QNode(circ_two, device_two)
  qml.set_shots(node_two, 1000)

  return node_two
```

```python
def make_logical_qbits(num_bits):
  def circ_one():
    for i in range(num_bits):
      j = i * 3
      qml.CNOT(wires = [j, j+1])
      qml.CNOT(wires = [j, j+2])
    for x in range(num_bits):
      y = x * 3
      qml.CNOT(wires = [y, y+1])
      qml.CNOT(wires = [y, y+2])
      qml.Toffoli(wires= [y+1, y+2, y])
    return qml.state()
  logical_qbit_device = qml.device("default.qubit", wires = num_bits * 3)

  node_one = qml.QNode(circ_one, logical_qbit_device)
  qml.set_shots(node_one, 1000)

  return node_one
```

```python
machine_one = make_logical_qbits(10)
print(qml.draw(machine_one)())
```
